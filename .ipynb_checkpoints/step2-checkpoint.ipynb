# Deep Dive into `soare`
Key insights:

- The second word depends entirely on the feedback from the first word
- The optimal next word need not come from the filtered set of candidates

```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from joblib import Parallel, delayed
from tqdm.notebook import tqdm

sns.set()
```

## Load Data

```python
with open('data/wordle-candidates.json', 'r') as file:
    wordle_candidates = json.load(file)
    
with open('data/wordle-answers.json', 'r') as file:
    wordle_answers = json.load(file)

wordle_candidates = pd.DataFrame(wordle_candidates['words'], columns=['word'])
wordle_answers = pd.DataFrame(wordle_answers['words'], columns=['word'])
wordle = wordle_candidates.loc[
    wordle_candidates.word.apply(lambda x: len(x)==len(set(x)))
].append(wordle_answers).reset_index(drop=True)
```

## Extract Letters

```python
# Extract letters
for i in range(5):
    wordle_candidates[f'l{i}'] = wordle_candidates.word.str[i]
    wordle_answers[f'p{i}'] = wordle_answers.word.str[i]
    wordle[f'p{i}'] = wordle.word.str[i]
```

```python
for l in list('abcdefghijklmnopqrstuvwxyz'):
    wordle_candidates[f'letter_{l}'] = wordle_candidates.word.apply(lambda x: x.count(l))
    wordle_answers[f'letter_{l}'] = wordle_answers.word.apply(lambda x: x.count(l))
    wordle[f'letter_{l}'] = wordle.word.apply(lambda x: x.count(l))
```

## Logic

```python
# Get GYX feedback from using `input_word` as a candidate, given that `solution` was the answer
def get_feedback(input_word, solution):
    output = ''
    for i in range(5):
        if input_word[i] == solution[i]:
            output += 'G'
        elif input_word[i] in solution:
            output += 'Y'
        else:
            output += 'X'
    return output
```

```python
def check_membership(feedback, candidate):
    if candidate == '':
        return np.nan
    newset = wordle.copy()
    for i, (letter, gyx) in enumerate(zip('soare', feedback)):
        if gyx == 'G':
            newset = newset.loc[newset[f'p{i}'] == letter]
        elif gyx == 'Y':
            newset = newset.loc[newset[f'letter_{letter}'].gt(0) & newset[f'p{i}'].ne(letter)]
        else:
            newset = newset.loc[newset[f'letter_{letter}'].eq(0)]
    return candidate in newset.word
```

```python
games['word_s2_in_cands'] = games.apply(lambda row: check_membership(row.feedback, row.word_s2), axis=1)
```

```python
games['word_s3_in_cands'] = games.apply(lambda row: check_membership(row.feedback, row.word_s3), axis=1)
```

## Compute Letter Frequencies

```python
def compute_letter_frequencies(wordset):
    w = wordset.copy()
    for letter in list('abcdefghijklmnopqrstuvwxyz'):
        w[letter] = w.word.str.contains(letter).astype(int)
    return w.iloc[:, 1:]

def compute_score(x, freqs):
    letters = set(x)
    output = 0
    for letter in letters:
        output += freqs[letter]
    return output
```

```python
lf_freqs = compute_letter_frequencies(wordle).sum().to_dict()
lf_scores = pd.DataFrame({'word': wordle.word,
                          'letter_freq': wordle.word.apply(compute_score, freqs=lf_freqs)})
```

## Compute Green/Yellow/Grey (GYX) Scores

```python
# Get GYX feedback from using `input_word` as a candidate, given that `solution` was the answer
def get_feedback(input_word, solution):
    output = ''
    for i in range(5):
        if input_word[i] == solution[i]:
            output += 'G'
        elif input_word[i] in solution:
            output += 'Y'
        else:
            output += 'X'
    return output
```

```python
# Get GYX scores
def get_scores(input_word):
    feedback = xxxyy.word.apply(lambda x: get_feedback(input_word, x))
    greens = feedback.apply(lambda x: x.count('G'))
    yellows = feedback.apply(lambda x: x.count('Y'))
    green_avg = greens.mean()
    yellow_avg = yellows.mean()
    weighted_avg = np.mean(greens * 3 + yellows)
    return input_word, green_avg, yellow_avg, weighted_avg
```

```python
raw_scores = Parallel(n_jobs=5, verbose=1)(delayed(get_scores)(input_word) for input_word in tqdm(wordle.word))
```

```python
df_scores = pd.DataFrame(raw_scores, columns=['word', 'green_avg', 'yellow_avg', 'weighted_avg'])
df_scores = df_scores.merge(lf_scores, how='left', on='word')
df_scores['green_rank'] = df_scores.green_avg.rank(ascending=False)
df_scores['yellow_rank'] = df_scores.yellow_avg.rank(ascending=False)
df_scores['wt_avg_rank'] = df_scores.weighted_avg.rank(ascending=False)
df_scores['lf_rank'] = df_scores.letter_freq.rank(ascending=False)
```

```python
shortlist = df_scores.loc[df_scores.green_rank.le(20) | \
                          df_scores.yellow_rank.le(20) | 
                          df_scores.wt_avg_rank.le(20) ]
                          # df_scores.lf_rank.le(20)]
```

## Compute No. of Candidates

```python
def compute_ncands(input_word, solution):
    fb = get_feedback(input_word, solution)
    newset = wordle_answers.loc[wordle_answers.word.isin(games.loc[games.feedback.eq('XXXYY'), 'solution'])].copy()
    for i, (letter, gyx) in enumerate(zip(input_word, fb)):
        if gyx == 'G':
            newset = newset.loc[newset[f'p{i}'] == letter]
        elif gyx == 'Y':
            newset = newset.loc[newset[f'letter_{letter}'].gt(0) & newset[f'p{i}'].ne(letter)]
        else:
            newset = newset.loc[newset[f'letter_{letter}'].eq(0)]
    return input_word, solution, fb, newset.shape[0]
```

```python
all_ncands = Parallel(n_jobs=4, verbose=1)(delayed(compute_ncands)(input_word, solution) for input_word in tqdm(shortlist.word) for solution in xxxyy.word)
```

```python
# Clean data
df_ncands = pd.DataFrame(all_ncands, columns=['word', 'solution', 'fb', 'ncands'])
```

```python
def entropy(x):
    probs = x.value_counts(normalize=True)
    return -np.sum(probs * np.log(probs))
```

```python
df_ncands = df_ncands.groupby('word').agg({'ncands': ['max', 'mean'],
                              'fb': [('nbuckets', pd.Series.nunique),
                                     ('bucket_entropy', entropy)]}).reset_index()
df_ncands.columns = ['word', 'ncands_max', 'ncands_mean', 'nbuckets', 'bucket_entropy']
```

## Combine Results

```python
df = shortlist.merge(df_ncands, how='left', on='word')
df['ncands_max_rank'] = df.ncands_max.rank()
df['ncands_mean_rank'] = df.ncands_mean.rank()
df['bucket_entropy_rank'] = df.bucket_entropy.rank(ascending=False)
df['avg_rank'] = df[['wt_avg_rank', 'ncands_max_rank', 'bucket_entropy_rank']].mean(axis=1)
```
